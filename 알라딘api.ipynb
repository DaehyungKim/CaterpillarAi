```python
import requests
import json
import time
# 갯수 뽑아오기
ALADIN_API_KEY = "ttbsk35021617001"
BASE_URL = "http://www.aladin.co.kr/ttb/api/ItemList.aspx"

all_titles = []

for start in range(1, 51, 50):
    params = {
        'ttbkey': ALADIN_API_KEY,
        'QueryType': 'Bestseller',
        'SearchTarget': 'Book',
        'MaxResults': 50,
        'start': start,
        'output': 'js',
        'CategoryId': 2551,
        'Version': '20131101'
    }

    try:
        response = requests.get(BASE_URL, params=params)
        response.raise_for_status()

        data = response.json()
        items = data.get('item', [])

        for book in items:
            title = book.get('author')
            if title:
                all_titles.append(title)

        print(f"{start}번째부터 {start+49}번째까지 처리 완료.")

        time.sleep(1)

    except requests.exceptions.RequestException as e:
        print(f"요청 오류: {e}")
    except json.JSONDecodeError:
        print("JSON 파싱 오류")
        print(f"Raw response: {response.text}")


print(f"\n총 {len(all_titles)}권 수집 완료:\n")
for i, title in enumerate(all_titles, 1):
    print(f"{i}. {title}")
```

```python
import re


def remove_parentheses(title):
    title = re.sub(r'\(.*?\)', '', title).strip()
    title = title.split(' - ')[0].strip()
    title = title.replace("'", "")
    title = re.sub(r'\s*초판본\s*', ' ', title).strip()
    return title

cleaned_titles = [remove_parentheses(title) for title in all_titles]
print(json.dumps(cleaned_titles, indent=4, ensure_ascii=False))
```
