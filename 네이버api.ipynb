```python
import requests
import json
import time

NAVER_CLIENT_ID = "CgZfTkZL3srUbinsP3lc"
NAVER_CLIENT_SECRET = "yazXm_z7WZ"
NAVER_BOOK_API_URL = "https://openapi.naver.com/v1/search/book.json"

def get_book_titles_simple(query, count=100):
    titles = []
    display = 100
    start = 1
    headers = {
        "X-Naver-Client-Id": NAVER_CLIENT_ID,
        "X-Naver-Client-Secret": NAVER_CLIENT_SECRET,
    }

    while len(titles) < count and start <= 1000:
        params = {
            'query': query,
            'display': min(display, count - len(titles)),
            'start': start
        }
        response = requests.get(NAVER_BOOK_API_URL, headers=headers, params=params)

        if response.status_code == 200:
            results = response.json()
            if 'items' in results and results['items']:
                fetched_items = results['items']
                for item in fetched_items:
                    title = item.get('title', '').replace('<b>', '').replace('</b>', '')
                    if title:
                        titles.append(title)
                        if len(titles) >= count:
                            break

                if len(fetched_items) < params['display'] or start + params['display'] > results.get('total', float('inf')):
                    break
                start += params['display']
                time.sleep(0.1)
            else:
                break
        else:
            print(f"API Error: Status Code {response.status_code}, start={start}")
            break

        if len(titles) >= count:
             break


    return titles[:count]

if __name__ == "__main__":
    search_term = "인공지능"
    book_titles = get_book_titles_simple(search_term, count=100)
    print(f"--- Collected {len(book_titles)} book titles for '{search_term}' ---")
    for i, title in enumerate(book_titles, 1):
        print(f"{i}. {title}")
```
